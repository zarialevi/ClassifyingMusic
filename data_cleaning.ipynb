```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.preprocessing import label_binarize

onehot = OneHotEncoder()
labelencoder = LabelEncoder()
```

```python
df_r = pd.read_csv('data/more_reggae_done.csv')
df_s = pd.read_csv('data/more_soca_done.csv')
df_d = pd.read_csv('data/more_dance_done.csv')
df_p = pd.read_csv('data/random_done.csv')
df_f = pd.read_csv('data/full_working.csv') 
df_m = pd.read_csv('data/fuller.csv')
```

```python
df_r['genre'] = 'reggae'
df_s['genre'] = 'soca'
df_d['genre'] = 'dancehall'
df_p['genre'] = 'pop'
```

```python
df = pd.concat([df_r, df_s, df_d, df_p, df_f, df_m], sort=True)
```

```python
df.drop_duplicates(subset='track_id', keep='first', inplace=True)
df.reset_index(inplace=True)
```

```python
df.to_csv('data/working_new.csv', index=True)
```

# Just features df

```python
df.info()
```

```python
feats = df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1','album_name', 'album_uri', 'analysis_url','type','album','album_id','artist','id','track_href','track_uri','track_name','artist_name','uri'])
```

```python
feats.info()
```

```python
feats.fillna(value=0, inplace=True)
```

```python
feats.columns
```

```python
year = []

for x in list(feats['release_date']):
    year.append(str(x)[:4])
    
feats['year'] = year    
feats['year'].astype('int64') 
feats.drop(columns='release_date', inplace=True)
```

```python
feats.info()
```

```python
df['genre'] = labelencoder.fit_transform(df['genre'])
df['explicit'] = labelencoder.fit_transform(df['explicit'])
df['key'] = labelencoder.fit_transform(df['key'])
df['mode'] = labelencoder.fit_transform(df['mode'])
df['time_signature'] = labelencoder.fit_transform(df['time_signature'])
```

```python
lst = ['key','mode','year', 'explicit','time_signature']

for x in lst:
    feats[x] = feats[x].astype('category')
```

```python
feats.info()
```

```python
df_dum = pd.get_dummies(feats[['key','mode','year', 'explicit','time_signature']])
```

```python
df_dum
```

```python
feats.drop(columns=['key','mode','year', 'explicit','time_signature'], inplace = True)
df = feats.join(df_dum)
```

```python
df.head()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
df.to_csv('data/feats.csv', index=False)
```
