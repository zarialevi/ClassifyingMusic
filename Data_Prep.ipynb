# Title

Once we have done all of API calls, we can concatenate the data, and remove and duplicates in case there has been any overlap.

0 = dancehall
1 = reggae
2 = soca
3 = Pop

```python

df_r = pd.read_csv('data/more_reggae_done.csv')
df_s = pd.read_csv('data/more_soca_done.csv')
df_d = pd.read_csv('data/more_dance_done.csv')
df_p = pd.read_csv('data/random_done.csv')
df_f = pd.read_csv('data/full_working.csv') 
df_m = pd.read_csv('data/fuller.csv')
```

```python
df_r['genre'] = 'reggae'
df_s['genre'] = 'soca'
df_d['genre'] = 'dancehall'
df_p['genre'] = 'pop'
```

```python
df = pd.concat([df_r, df_s, df_d, df_p, df_f, df_m], sort=True)
```

```python
df.drop_duplicates(subset='track_id', keep='first', inplace=True)
df.fillna(value=0, inplace=True)
df.reset_index(inplace=True)
```

```python
# df.to_csv('data/working.csv', index=False)
```

```python
df.fillna(value=0, inplace=True)
```

```python
year = []

for x in list(df['release_date']):
    year.append(str(x)[:4])
    
df['year'] = year    
df['year'].astype('int64') 
df.drop(columns='release_date', inplace=True)
```

```python
df['genre'] = labelencoder.fit_transform(df['genre'])
df['explicit'] = labelencoder.fit_transform(df['explicit'])
df['key'] = labelencoder.fit_transform(df['key'])
df['mode'] = labelencoder.fit_transform(df['mode'])
df['time_signature'] = labelencoder.fit_transform(df['time_signature'])
```

```python
lst = ['key','mode','year', 'explicit','time_signature']

for x in lst:
    df[x] = df[x].astype('category')
```

```python
df_dum = pd.get_dummies(df[['key','mode','year', 'explicit','time_signature']])
```

```python
df.drop(columns=['key','mode','year', 'explicit','time_signature'], inplace = True)
df = df.join(df_dum)
```
