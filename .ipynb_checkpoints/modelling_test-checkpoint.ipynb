```python
import pandas as pd
import numpy as np

from pandas_profiling import ProfileReport

import matplotlib.pyplot as plt
from matplotlib import rc

import seaborn as sns

from random import sample
```

```python
df = pd.read_csv('full_working.csv')
```

```python
df['genre'] = df['genre'].astype('category')
```

```python
df.head()
```

```python
profile = ProfileReport(df, title='Pandas Profiling Report', style={'full_width':True})
profile
```

```python
col_df = pd.DataFrame(df.groupby(['genre', 'explicit']).size().unstack())
col_df.plot(kind='bar', stacked = True, colormap='Greens')
plt.legend(bbox_to_anchor = [1.25,1], title='Key', loc=1, fontsize='medium')
```

```python
col_df = pd.DataFrame(df.groupby(['genre', 'key']).size().unstack())
col_df.plot(kind='bar', stacked = True, colormap='Greens', legend=True)
plt.legend(bbox_to_anchor = [1.25,1], title='Key', loc=1, fontsize='medium');
```

```python
df2 = df.sample(1250, random_state=123)

sns.lmplot(x = 'tempo', y = 'energy', data=df2, fit_reg=False, hue='genre', palette='Greens', legend=False);
plt.legend(loc='bottom right');
```

```python
sns.lmplot(x = 'danceability', y = 'tempo', data=df2, fit_reg=False, hue='genre', palette='Greens', legend=False);
plt.legend(loc='bottom right');
```
