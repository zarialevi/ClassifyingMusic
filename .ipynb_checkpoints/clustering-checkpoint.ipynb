```python
import pandas as pd
import numpy as np

from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
from sklearn.cluster import AgglomerativeClustering
from sklearn.neighbors import KernelDensity
import matplotlib.pyplot as plt
np.random.seed(0)
%matplotlib inline  
```

```python
df = pd.read_csv('data/working_new.csv')
```

```python
df.columns
```

```python
df.drop(columns=['Unnamed: 0', 'album_id', 'artist', 'artist_name','explicit', 'id','release_date','track_href','track_id', 'track_name', 'track_uri', 'uri'], inplace=True)
df.dropna(inplace=True)
```

```python
y = df['genre']
X = df.drop(columns=['genre'], axis=1)
```

```python
agg_clust = AgglomerativeClustering(n_clusters=10)
assigned_clust = agg_clust.fit_predict(X)
```

```python
plt.scatter(df.loudness, df.energy, c=assigned_clust, s=3);
```

```python
from sklearn.cluster import KMeans
```

```python
k_means = KMeans(n_clusters = 7)
k_means.fit(X)
y_hat = k_means.predict(X)
```

```python
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c = y_hat, s = 25)
cl_centers = k_means.cluster_centers_
plt.scatter(cl_centers[:, 0], cl_centers[:, 1], c='black', s=100);
```
