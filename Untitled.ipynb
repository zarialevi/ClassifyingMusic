```python
from sklearn.manifold import TSNE
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
labelencoder = LabelEncoder()
```

```python
df = pd.read_csv('data/working_new.csv')
```

```python
df2 = df.drop(columns=['Unnamed: 0', 'album_id', 'artist', 
                 'artist_name','id','track_name','uri',
                 'release_date','track_href','track_uri',
                'track_id','genre'])

df2.dropna(inplace=True)
```

```python
df2.columns
```

```python
df['genre'] = labelencoder.fit_transform(df['genre'])
df['explicit'] = labelencoder.fit_transform(df['explicit'])
```

```python
df.drop(columns=['tsne_1','tsne_2'], inplace=True)
```

```python
tsne = TSNE(n_components=2, verbose=0, perplexity=200, learning_rate=200, n_iter=5000, init='pca')
```

```python
tsne_results = tsne.fit_transform(df2)
```

```python
tsne.n_iter_
```

```python
df2['tsne_1'] = pd.Series(tsne_results[:,0])
df2['tsne_2'] = pd.Series(tsne_results[:,1])
df2['genre'] = df['genre']
df['genre'] = labelencoder.fit_transform(df['genre'])
```

```python
plt.figure(figsize=(16,10))
sns.scatterplot(x="tsne_1", y="tsne_2",
                hue="genre",
                data=df,
                legend="full",
                palette="RdBu",
                alpha=0.8)
```
