```python
from pyAudioAnalysis import audioBasicIO
from pyAudioAnalysis import ShortTermFeatures
from pyAudioAnalysis import MidTermFeatures
from pyAudioAnalysis import audioVisualization
import matplotlib.pyplot as plt
%matplotlib inline
%load_ext autoreload
%reload_ext autoreload
```

```python
from pyAudioAnalysis import convertToWav

convertToWav.convert_dir_mp3_to_wav(audio_folder='music/samples', sampling_rate, num_channels, use_tags=False)

sam_rate, signal = audioBasicIO.read_audio_file("music/samples/scottish.wav")
```

```python
sam_rate, signal = audioBasicIO.read_audio_file("pyAudioAnalysis/data/scottish.wav")
```

```python
from pyAudioAnalysis import ShortTermFeatures

F, f_names = ShortTermFeatures.feature_extraction(signal, sam_rate, window=sam_rate*0.050, step=sam_rate*0.025)
```

```python
for x in range(0,len(f_names)):
    plt.subplot(2,1,1); plt.plot(F[x,:]); plt.xlabel('Frame no'); plt.ylabel(f_names[x]);plt.show()
```

```python
audioVisualization.visualizeFeaturesFolder('samples', dimReductionMethod=None)
```

```python


mid_features, short_features, mid_feature_names = MidTermFeatures.mid_feature_extraction(signal=signal, sampling_rate=sam_rate, 
                                       mid_window = sam_rate*0.5, mid_step = sam_rate*0.5,
                                       short_window = sam_rate*0.050, short_step=sam_rate*0.025)
```

```python
for x in range(0,34):
    plt.subplot(2,1,1); plt.plot(mid_features[x,:]); plt.xlabel('Frame no'); 
    plt.ylabel(mid_feature_names[x]);plt.show()
```

```python
chromogram, time_axis, freq_axis = ShortTermFeatures.chromagram(signal=signal, sampling_rate=sam_rate, window=sam_rate*0.50, step=sam_rate*0.25, plot=True) 
```

```python
spectrogram, time_axis, freq_axis = ShortTermFeatures.spectrogram(signal=signal, sampling_rate=sam_rate, window=sam_rate*0.50, 
                             step=sam_rate*0.25, plot=True) 
```
