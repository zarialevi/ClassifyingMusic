```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.preprocessing import label_binarize
```

```python
onehot = OneHotEncoder()
labelencoder = LabelEncoder()
```

```python
df = pd.read_csv('data/random_full.csv')
```

```python
df['genre'] = 'Pop'
```

```python
df.columns
```

```python
df['genre'] = labelencoder.fit_transform(df['genre'])
df['explicit'] = labelencoder.fit_transform(df['explicit'])
df['key'] = labelencoder.fit_transform(df['key'])
df['mode'] = labelencoder.fit_transform(df['mode'])
df['time_signature'] = labelencoder.fit_transform(df['time_signature'])
```

```python
df.drop(columns = ['Unnamed: 0', 'type','album','artist','track_name'], inplace=True)
```

```python
year = []

for x in df['release_date']:
    year.append(x[:4])
    
df['year'] = year    
df['year'].astype('int64') 
df.drop(columns='release_date', inplace=True)
```

```python
lst = ['key','mode','year', 'explicit','time_signature']

for x in lst:
    df[x] = df[x].astype('category')
```

```python
df_dum = pd.get_dummies(df[['key','mode','year', 'explicit','time_signature']])
```

```python
df_dum
```

```python
df.drop(columns=['key','mode','year', 'explicit','time_signature'], inplace = True)
df = df.join(df_dum)
```

```python
df.head()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
df.drop(columns=['Unnamed: 0.1','track_uri'], inplace=True)
```

```python
df.to_csv('data/random_clean.csv', index=False)
```
